# Detection Family Figure

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import json

plt.style.use("./style_sheet.mplstyle")

col = ["xkcd:muted blue", "xkcd:coral pink"]
```

```python
# Load data
fname = "figdata/ideal_data_detection_task_family.json"
data = json.load(open(fname, "r"))
s_range = data["s_range"]
accuracies = data["accuracies"]
```

```python
# Figure
plt.subplots(constrained_layout=True)
for pairs in [False, True]:
    plt.plot(
        s_range[::-1],
        np.array(accuracies[str(pairs).lower()]) * 100,
        color=col[pairs],
        linewidth=5.0,
        label="FtA" if pairs else "AtF",
    )
plt.legend(loc="best")

plt.xticks([0, 1], ["Dense", "Sparse"])
plt.xlabel("Signal Sparsity")
plt.ylim([65, 85])
plt.ylabel("Accuracy")
plt.savefig("detection_family.svg")
```
